```python
from P2D import P2D

timerange=['2022-06-01', '2022-06-21T06:00']

model = P2D.P2D(input_data=['OMNI', 'STEREO-A', 'PSP', 'SolO']
                , timerange=timerange
                , propagation_time='2d'
                , model_type='ballistic')#, coordinate_system='other')

model.plot(rlim=2)#, variable_to_plot='Sim_step')
#plot_heliosphere.plot_spacecraft_carrington(model.model, rlim=5)#, variable_to_plot='B')
```

```python
cutout = model.observe_from('OMNI')
```

```python
import matplotlib.pyplot as plt

model.plot_timeseries(model=cutout)
plt.show()
```
